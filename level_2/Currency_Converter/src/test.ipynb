```python
import requests
```

```python
url = "https://api.exchangerate-api.com/v4/latest/USD"
```

```python
rg = requests.get(url)
```

```python
rg.json()['rates']
```

```python
base_currency = 'USD'
target_currency = 'CAD'


def exchange_rates(base_currency, target_currency):
    base_cr = rg.json()['rates'][base_currency]
    target_cr = rg.json()['rates'][target_currency]
    print(f"1 {base_currency} = {target_cr/base_cr:.4f} {target_currency}")


exchange_rates(base_currency='XPF', target_currency=target_currency)
```
